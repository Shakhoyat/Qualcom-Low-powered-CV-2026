```python
import qai_hub as hub

device = hub.Device("Dragonwing IQ-9075 EVK")

compile_job = hub.submit_compile_job(
    model="C:\\Users\\USER\\Downloads\\Telegram Desktop\\model.onnx",   # your file
    device=device,
    input_specs={
        "input": (1, 3, 16, 112, 112)   # ⚠️ adjust this
    },
    options="--target_runtime tflite"
)

target_model = compile_job.get_target_model()
```

```python
import numpy as np

# Match your input shape
dummy_input = np.random.rand(1, 3, 16, 112, 112).astype("float32")

job = hub.submit_inference_job(
    model=target_model,
    device=device,
    inputs={"input": [dummy_input]}
)

output = job.download_output_data()
print(output)
```

```python
profile_job = hub.submit_profile_job(
    model=target_model,
    device=device
)
```
